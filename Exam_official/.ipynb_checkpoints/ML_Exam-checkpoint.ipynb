### 1. Data reading and exploration.

```python
edu_data = pd.read_excel("data/Data_User_Modeling_Dataset_Hamdi Tolga KAHRAMAN.xls")
edu_data.head(3)
```

Short description of what each column means:
* STG (The degree of study time for goal object materails), (input value) 
* SCG (The degree of repetition number of user for goal object materails) (input value) 
* STR (The degree of study time of user for related objects with goal object) (input value) 
* LPR (The exam performance of user for related objects with goal object) (input value) 
* PEG (The exam performance of user for goal objects) (input value) 
* UNS (The knowledge level of user) (target value) 

```python
edu_data.info()
```

The dataset contains no missing data, five number of attributes (four integer and 1 categorical), 403 number of Instances.

```python
edu_data["UNS"].value_counts()
```

To be able to modeling the data, last column which is my target have to convert to integer. By grouping by target i notice mistake in category "Very Low" and "very_low".
By condition count of target distribution is: Very Low: 50, Low:129, Middle: 122, High 130

```python
edu_data['UNS'] = edu_data['UNS'].map({'very_low': 1, 'Very Low': 1, 'Low': 2, 'Middle': 3, 'High': 4})
```

```python
edu_data.UNS.value_counts()
```

```python
edu_data.mean()
```

```python
scatter_matrix(edu_data, alpha=0.2, figsize=(6, 6), diagonal='kde')
plt.show()
```

### 2. Data modeling.
The idea is to chose best model for prediction from the next:
* LnearRegresion
* Decision Tree
* Random Forest
* Linear SVM
* Gaussian SVM

The dataset is small and that will allow me to test thes different model and to tune there hyper parameters.

#### Procedure for processing:
* Leave 30% from data for testing.
* Perform cross_validation on the rest 70%.
* Fine tune the model with GridSearchCV and hyper parameters.
* Run the best model on the test data.


The variables are saved in `attributes` and target in `labels`.

```python
attributes = edu_data.drop("UNS", axis=1)
labels = edu_data.UNS
print(edu_data.shape)
print(attributes.shape)
print(labels.shape)
```

Use the standard 70% / 30% split. The split is stratify according to the `labels`.

```python
attributes_train, attributes_test, labels_train, labels_test = \
train_test_split(attributes, labels, train_size = 0.7, test_size = 0.3, stratify=labels)
print(attributes_train.shape)
print(labels_train.shape)
```

## Logistic Regresion

Use grid search with the following grid:

* C: 0.01, 0.1, 1, 4, 5
* penalty: l1, l2

#### 1. Tune model

```python
tuned_params = [{"C": [0.01, 0.1, 1, 4, 5], "penalty": ['l1', 'l2']}]

grid_lr = GridSearchCV(LogisticRegression(),tuned_params, cv=5)
grid_lr.fit(attributes_train, labels_train)
print(f"Best Params: {grid_lr.best_params_}")
```

#### 2. Cross validation score on the training data.

```python
logistic_regresion_cv_score = cross_val_score(grid_lr.best_estimator_, attributes_train, labels_train, cv = 5)
print(f"cross_validation_scores: {logistic_regresion_cv_score}")
print("mean_cv_score: %0.2f (+/- %0.2f)" % (logistic_regresion_cv_score.mean(),
                                                       logistic_regresion_cv_score.std() * 2))
```

#### 3. Run the model on the test data.

```python
print("Logistic_regresion: %0.2f" % (grid_lr.best_estimator_.score(attributes_test, labels_test)))
```

## Decision Tree

Use grid search with the following grid:

max_depth: 1, 5, 7, 15, 20
min_samples_leaf: 1, 2, 3
max_leaf_nodes: 6, 7, 8, 9

#### 1. Tune model

Use grid search with the following grid:
* max_depth: 1, 5, 7, 15, 20
* min_samples_leaf: 2, 5, 10, 12
* max_leaf_nodes: 5, 10, 20

```python
tuned_params = [{"max_depth": [1, 5, 7, 15, 20], "min_samples_leaf": [1, 2, 3], "max_leaf_nodes": [6, 7, 8, 9] }]

grid_tree = GridSearchCV(DecisionTreeClassifier(),tuned_params, cv = 5)
grid_tree.fit(attributes_train, labels_train)
print(f"Best Params: {grid_tree.best_params_}")
```

#### 2. Cross validation score on the training data.

```python
decision_tree_cv_score = cross_val_score(grid_tree.best_estimator_, attributes_train, labels_train, cv = 5)
print(f"cross_validation_scores: {decision_tree_cv_score}")
print("mean_cv_score: %0.2f (+/- %0.2f)" % (decision_tree_cv_score.mean(),
                                                       decision_tree_cv_score.std() * 2))
```

#### 3. Run the model on the test data.

```python
print("Desicion tree: %0.2f" % (grid_tree.best_estimator_.score(attributes_test, labels_test)))
```

##  Random Forest

#### 1. Tune model

Use the following grid:
* n_estimators: 38, 40, 42, 45 
* max_depth: 4, 5, 6, 7

```python
tuned_params = [{"n_estimators": [38, 40, 42, 45], "max_depth": [4, 5, 6, 7]}]

grid_forest = GridSearchCV(RandomForestClassifier(),tuned_params, cv = 5)
grid_forest.fit(attributes_train, labels_train)
print(f"Best Params: {grid_forest.best_params_}")
```

#### 2. Cross validation score on the training data.

```python
random_forest_cv_scores = cross_val_score(grid_forest.best_estimator_, attributes_train, labels_train, cv = 5)
print(f"cross_validation_scores: {random_forest_cv_scores}")
print("mean_cv_score: %0.2f (+/- %0.2f)" % (random_forest_cv_scores.mean(),
                                                       random_forest_cv_scores.std() * 2))
```

#### 3. Run the model on the test data.

```python
print("Random forest: %0.2f" % (grid_forest.best_estimator_.score(attributes_test, labels_test)))
```

##  Linear SVM

#### 1.Tune model.

Use the following grid:
* C: 6,7,8.5,9

```python
tuned_params = [{"C": [6,7,8.5,9]}]

grid_svm = GridSearchCV(LinearSVC(),tuned_params, cv = 5)
grid_svm.fit(attributes_train, labels_train)
print(f"Best Params: {grid_svm.best_params_}")
```

#### 2. Cross validation score on the training data.

```python
linear_svm_cv_scores = cross_val_score(grid_svm.best_estimator_, attributes_train, labels_train, cv = 5)
print(f"cross_validation_scores: {linear_svm_cv_scores}")
print("mean_cv_score: %0.2f (+/- %0.2f)" % (linear_svm_cv_scores.mean(),
                                                       linear_svm_cv_scores.std() * 2))
```

#### 3. Run the model on the test data.

```python
print("Linear SVM: %0.2f" % (grid_svm.best_estimator_.score(attributes_test, labels_test)))
```

## Gaussian SVM

#### 1. Tune model

* C: 40, 48, 50, 60
* gamma: 0.001, 0.01, 0.1, 0.2, 0.3, 0.4

```python
tuned_params = [{"C": [40, 48, 50, 60], "gamma": [0.001, 0.01, 0.1, 0.2, 0.3, 0.4]}]

grid_gaussian = GridSearchCV(SVC(),tuned_params, cv=5)
grid_gaussian.fit(attributes_train, labels_train)
print(f"Best Params: {grid_gaussian.best_params_}")
```

#### 2. Cross validation score on the training data.

```python
gaussian_svm_cv_scores = cross_val_score(grid_gaussian.best_estimator_, attributes_train, labels_train, cv = 5)
print(f"cross_validation_scores: {gaussian_svm_cv_scores}")
print("mean_cv_score: %0.2f (+/- %0.2f)" % (gaussian_svm_cv_scores.mean(),
                                                       gaussian_svm_cv_scores.std() * 2))
```

#### 3. Run the model on the test data.

```python
print("Gassian SVM: %0.2f" % (grid_gaussian.best_estimator_.score(attributes_test, labels_test)))
```
